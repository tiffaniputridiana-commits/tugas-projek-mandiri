Soal 3 lakukan tahapan pada soal 1 dengan menggunakan PyTorch

```python
import torch as pt
```

```python
# buat variabel skalar x dan simpan suatu nilai skalar
x_pt = 10
x_pt
```

```python
# tampilkan tipe data dari x
x_pt = pt.tensor(10, dtype=pt.int16)
x_pt
```

```python
# buat variabel skalar y dan simpan suatu nilai skalar
y_pt = 10
y_pt
```

```python
# tampilkan tipe data dari y
y_pt = pt.tensor(10, dtype=pt.int16)
y_pt
```

```python
# jumlahkan x+y
x_pt + y_pt
```

```python
# buat variabel bernama x_float dan simpan suatu nilai desimal
xpt_float = 10.0
```

```python
# jumlahkan x_float dengan y
ptfloat_sum = pt.add(xpt_float, y_pt)
ptfloat_sum
```

```python
xpt_float = pt.tensor(10.0, dtype=pt.float16)
xpt_float
```
